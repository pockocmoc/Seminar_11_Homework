```python
from sympy import *
from sympy.plotting import plot as plt

init_printing()

x = Symbol('x')
fx = sin(x) ** 2 - cos(x) ** 2
```

```python
x1, x2 = solve(fx)
print('первый корень', round(x1.evalf(), 2), '\nвторой корень', round(x2.evalf(), 2))
```
